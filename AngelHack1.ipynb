```python
schools = []
scores = []
teach = []
outlook = []

for i in df1['name']:
    schools.append(i)

for i in df1['scores_overall']:
    scores.append(i)
    
for i in df1['scores_teaching']:
    teach.append(i)
    
for x in df1['scores_international_outlook']:
    try:
        outlook.append(float(x))
    except:
        outlook.append(50)
        pass

sf = [schools, scores, teach, outlook,]
```

```python
plt.plot(scores[:30], 'b', marker='*', label='Student overall scores')
plt.plot(teach[:30], 'r', marker= '*', label='Teaching performance')
plt.plot(outlook[:30], 'y', marker='*', label='International Outlook')
plt.legend()
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(18, 6))
ax.barh([i+1 for i in range(5)], scores[:5], color='#dd2222')

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
    
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 10)
ax.yaxis.set_tick_params(pad = 10)

# Add x, y gridlines
ax.grid(b = True, color ='#222222',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.3)

ax.set_ylabel(ax.get_ylabel(), fontdict={'fontsize':20})
ax.set_xlabel(ax.get_ylabel(), fontdict={'fontsize':24})
ax.bar_label(ax.containers[0], labels=schools[:5], label_type='center', fontsize=20, fontweight='bold', color='#ffffff')

# Show top values
ax.invert_yaxis()

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 15, fontweight ='light',
             color ='#000000')

# Add Plot Title
ax.set_title('Top 10 schools and their comparison in student scores',
             loc ='left', fontsize=25)
```
